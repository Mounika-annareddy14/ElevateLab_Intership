# 1. `Identifying and handling missing values`

### - `Only income column has missing values`

```python
# check how many are missing
print(dataset.isnull().sum())


# replacing null values with median (because median is good for skewed data)
dataset["Income"] = dataset["Income"].fillna(dataset["Income"].median())

# checking for missing values
dataset.isnull().sum()
```

# 2. Removing Duplicates



### a. `check for duplicate first`

```python
duplicted_count = dataset.duplicated().sum()
print(f"Number of duplicate rows: {duplicted_count}")
```

### - `We have 182 duplicate rows now we are dropping them`

```python
dataset = dataset.drop_duplicates()
```

```python
# re-check
print(f"Checking the shape of dataset after dropping duplicates {dataset.shape}")
```

### -`Previously we have 2240 rows now we have 2058 rows`

# 3.`Standardize text values`

### a.`Check unique values before cleaning`

```python
print(dataset['Education'].unique())
print(dataset['Marital_Status'].unique())
```

### `b.remove whitespace and apply title case`

```python
dataset['Education'] = dataset['Education'].str.strip().str.title()
dataset['Marital_Status'] = dataset['Marital_Status'].str.strip().str.title()
```

### `c.Grouping similar values`

```python
dataset['Marital_Status'] = dataset['Marital_Status'].replace({
    'Alone':'Single',
    'YOLO':'Single',
    'Absurd':'Single',
    'Widow':'Widowed',
    'Yolo':'Single'
})
```

### `d.Recheck for Unique values`

```python
print(dataset['Education'].unique())
print(dataset['Marital_Status'].unique())
```

# `4.Convert Dates to a Consistent Format`

### `a.Check the current type and sample values`

```python
print(dataset['Dt_Customer'].dtype)
print(dataset['Dt_Customer'].head())
```

### - `It is an object type as we know date should be in dd-mm-year format`

```python
dataset['Dt_Customer'] = pd.to_datetime(dataset['Dt_Customer'], format='%d-%m-%Y')
```

### `b.recheck`

```python
print(dataset['Dt_Customer'].dtype)
print(dataset['Dt_Customer'].head())
```

# `5.Clean Column names`

```python
dataset.columns = dataset.columns.str.strip()  #-->removing leading and trailing spaces
dataset.columns = dataset.columns.str.lower() #--> lower casing the columns
dataset.columns = dataset.columns.str.replace(' ' , '_')  #-->converting spaces with underscores
dataset.columns = dataset.columns.str.replace('[^a-z0-9_]' , '' , regex=True)  # remove special characters
```

```python
print(dataset.columns)
```

# `6.Check and fix datatypes`

### `a.Check current datatypes:`

```python
print(dataset.dtypes)
```

### `b.Create and fix age column instead of birth year`

```python
dataset['age'] = 2025 - dataset['year_birth']
dataset['age'] = dataset['age'].astype(int)
```

### `C.Income column as float`

```python
dataset['income'] = dataset['income'].astype(float)
```

### `d.make sure binary column as int`

```python
binary_columns = ['complain', 'response', 'acceptedcmp1', 'acceptedcmp2', 'acceptedcmp3',
                  'acceptedcmp4', 'acceptedcmp5']
dataset[binary_columns] = dataset[binary_columns].astype(int)
```
